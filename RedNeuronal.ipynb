```python
import tensorflow as tf
import numpy as np
```

```python
"""
Declarar un arreglo de numeros de celsius y una arreglo de numero Fahrenheit
son los ejemplos
"""

#tuplas te permite tener diferentes tipos de datos
celsius= np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit = np.array([-40,14,32,46,59,72,100], dtype=float)
```

```python
#Usar framework Keras (Keras esta en tf)
#capas es 1, y se pone con 1, y la  entrada le pusimos 1
capa= tf.keras.layers.Dense(units=1, input_shape=[1])
modelo= tf.keras.Sequential([capa])
```

```python
#el 0.1 es la referencia de peso y cesgo, si colocamos un numero grande pasara un numero que no podra hacer el aprendizaje de manera adecuada
#hace cambios pequeños, y es mejor uno pequeños a que se equivoque en grandes
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
#indicar el entrenamiento
#feed le indica los datos esperado y le alimentamos
# 3 datos es cuantas veces debe de optimizar, o cuanto tuvo que practicar
# verbose, para saber en que punto llego su aprendizase y lo demas ya tiene cambios minimo, y se puede valida con una grafica, mostrara cuanto va el progreso
#cada ejecuccion de esto es una capa, como corri y cancele, la grafica se mira raro
print("Iniciar entrenamiento")
historial= modelo.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print("Modelo ha entrenado!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de Perdida")
plt.plot(historial.history['loss'])
```

```python
print("La predicción...")
entrada= np.array([15])
resultado= modelo.predict(entrada)
print("El resultado es "+str(resultado)+" fahnrenheit!")
print("\nVariables internas del modelo...")
print(capa.get_weights())
```
